## Setup

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## Estadisticas por fuera de los datos

### Patentes adulteradas

```python
def plot_patentes_adulteradas():
    participants = ["Chiche"]
    patentes_adulteradas_found = [1]

    # Function to conditionally display the number
    def autopct(pct):
        absolute = round(pct / 100.0 * sum(patentes_adulteradas_found))
        return str(absolute) if pct > 0 else ""

    # Plot the number of adulterated patents found by each participant
    plt.pie(
        patentes_adulteradas_found,
        labels=participants,
        autopct=autopct,
        colors=sns.color_palette("crest", len(participants)),
    )

    plt.title("N√∫mero de patentes adulteradas encontradas")
    plt.show()

plot_patentes_adulteradas()
```

### Pifies de mandar la patente incorrecta

```python
def plot_pifies():
    participants = ["Chiche", "Germo"]
    patentes_adulteradas_found = [1, 1]

    # bar chart
    plt.bar(participants, patentes_adulteradas_found, color=sns.color_palette("crest", len(participants)))

    # Put the number of patents found on top of the bars
    biggest = max(patentes_adulteradas_found)
    space_between = biggest * 0.01

    for i in range(len(participants)):
        plt.text(
            i,
            patentes_adulteradas_found[i] + space_between,
            patentes_adulteradas_found[i],
            ha="center",
        )

    plt.title("Cantidad de pifies realizados")
    plt.show()


plot_pifies()
```
